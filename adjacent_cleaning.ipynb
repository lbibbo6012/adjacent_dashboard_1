```python
import json
import pandas as pd
```

```python
follows = []
with open('export_All-Follows_2025-02-21_16-47-10.ndjson') as f:
    for line in f:
        follows.append(json.loads(line))

print(follows)
```

```python
follows_df = pd.DataFrame(follows)
```

```python
follows_df.head()
```

```python
follows_df['Slug'].unique()
```

```python
follows_df['Creator'].value_counts()
```

#### Drop the Slug Column Since it's Blank

```python
follows_df.drop(labels='Slug', axis=1, inplace=True)
```

```python
follows_df.info()
```

#### Change Creation Date and Modified Data Columns to Datetimes

```python
follows_df['Creation Date'] = pd.to_datetime(follows_df['Creation Date'])
follows_df['Modified Date'] = pd.to_datetime(follows_df['Modified Date'])
follows_df.info()
```

```python
posts = []
with open('export_All-Posts_2025-02-21_16-50-21.ndjson') as f:
    for line in f:
        posts.append(json.loads(line))

print(posts)
```

```python
posts_df = pd.DataFrame(posts)
```

```python
posts_df.head()
```

#### Change 'Interactions' and 'Comment' to Integer Type

```python
posts_df['Interactions'].value_counts()
```

```python
posts_df['Interactions'] = posts_df['Interactions'].astype(int)
```

```python
posts_df['Comment'].value_counts()
```

```python
#replace '' with 0 first in the Comment column
posts_df['Comment'].replace('', 0, inplace=True)
```

```python
posts_df['Comment'] = posts_df['Comment'].astype(int)
```

```python
posts_df['Comment'].value_counts()
```

#### Convert the Downvote and Upvote Column to a List of Usernames

```python
posts_df['Upvote'] = posts_df['Upvote'].apply(lambda x: x.split(','))
```

```python
posts_df['Downvote'] = posts_df['Downvote'].apply(lambda x: x.split(','))
```
